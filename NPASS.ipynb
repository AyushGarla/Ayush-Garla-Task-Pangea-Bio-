```python
import pandas as pd
import numpy as np
compound_df = pd.read_csv("F:\\MS\\SRH study\\IV sem\\INT\\Datasets 2023\\Natural Products-General Information.txt", sep='\t')
compound_df.head()
```

Data cleaning
--

```python
compound_df.isnull().sum()
```

```python
compound_df[compound_df.isnull().any(axis=1)] #This data can be removed or replaced according to the requirment, as of now i will be removing it.
```

```python
compound_df=compound_df.dropna().reset_index(drop=True)
```

```python
compound_df.replace('n.a.', np.nan, inplace=True)
compound_df
```

```python
compound_df.isnull().sum() #checking
```

```python
for val in compound_df:
    compound_df[val].fillna('Missing',inplace=True)
compound_df.head(50)
```

```python
compound_df.isnull().sum() #checking
```

```python
# Checking for duplicate np_id values
duplicates = compound_df[compound_df.duplicated(subset='np_id', keep=False)]
if not duplicates.empty:
    print("Duplicate np_id values found:")
    print(duplicates)
else:
    print("No duplicate np_id values found.")
```

Data formatting
--

```python
compound_df.columns = [col.strip().lower() for col in compound_df.columns] #keeping column names standard
```

```python
compound_df.columns
```

```python
compound_df.info()
```

Feature Engineering
--

```python
compound_df['if_has_quantity_binary'] = compound_df['if_has_quantity'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
compound_df = compound_df.drop(columns=['if_has_quantity'])
compound_df = compound_df.rename(columns={'if_has_quantity_binary': 'if_has_quantity'})
compound_data=compound_df
compound_data
```

```python
# Checking for duplicate np_id values
duplicates = Structure_df[Structure_df.duplicated(subset='np_id', keep=False)]
duplicate_count = duplicates['np_id'].nunique()  # Count unique duplicate np_id values

if not duplicates.empty:
    print("Duplicate np_id values found:")
    print(duplicates)
    print(f"Number of duplicate np_id values: {duplicate_count}")
else:
    print("No duplicate np_id values found.")

# Remove duplicate np_id values
Structure_df = Structure_df.drop_duplicates(subset='np_id', keep='first')

Structure_df
```

```python
import psycopg2
from sqlalchemy import create_engine
import pandas as pd

# Connecting to PostgreSQL database
connection = psycopg2.connect(
    host="localhost",
    database="compound",
    user="postgres",
    password="123456"
)

# Creating an engine for SQL Alchemy
engine = create_engine('postgresql://postgres:123456@localhost/compound')

# DataFrame to a SQL table
compound_data.to_sql('npass_compound_table', engine, if_exists='replace', index=False)
structure_data.to_sql('npass_structure_table', engine, if_exists='replace', index=False)
activity_data.to_sql('npass_activity_table', engine, if_exists='replace', index=False)
Species_data.to_sql('npass_species_table', engine, if_exists='replace', index=False)
target_data.to_sql('npass_target_table', engine, if_exists='replace', index=False)
source_data.to_sql('npass_source_table', engine, if_exists='replace', index=False)

# Close the connection
connection.close()

print("DataFrame successfully written to PostgreSQL table.")
```
