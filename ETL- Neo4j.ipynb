```python

from neo4j import GraphDatabase

uri = "bolt://localhost:7687" 
username = "neo4j"  # 
password = "Pan@1234"  

# Connecting to the Neo4j database
driver = GraphDatabase.driver(uri, auth=(username, password))

#nodes for compound
def create_nodes(tx, np_id, pref_name, iupac_name, chembl_id, pubchem_cid):
    tx.run(
        "CREATE (n:Compound {np_id: $np_id, pref_name: $pref_name, iupac_name: $iupac_name, chembl_id: $chembl_id, pubchem_cid: $pubchem_cid})",
        np_id=np_id,
        pref_name=pref_name,
        iupac_name=iupac_name,
        chembl_id=chembl_id,
        pubchem_cid=pubchem_cid
    )

for index, row in compound_data.iterrows():
    with driver.session() as session:
        session.write_transaction(create_nodes, row['np_id'], row['pref_name'], row['iupac_name'], row['chembl_id'], row['pubchem_cid'])
#########################################
# nodes for structure
def create_structure_nodes(tx, np_id, inchi, inchikey, smiles):
    tx.run("CREATE (:Structure {np_id: $np_id, inchi: $inchi, inchikey: $inchikey, smiles: $smiles})", 
           np_id=np_id, inchi=inchi, inchikey=inchikey, smiles=smiles)

with driver.session() as session:
    for index, row in Structure_data.iterrows():
        session.write_transaction(create_structure_nodes, row['np_id'], row['inchi'], row['inchikey'], row['smiles'])
####################################################################
#nodes for uses
# Function to create nodes
def create_uses_nodes(tx, id,species, tribe, source, page_no, use_category, use_subcategory, notes, raw_source):
    tx.run(
        "CREATE (:Usage {id: $id,species: $species, tribe: $tribe, source: $source, page_no: $page_no, use_category: $use_category, use_subcategory: $use_subcategory, notes: $notes, raw_source: $raw_source})",
        id=id,species=species, tribe=tribe, source=source, page_no=page_no, use_category=use_category, use_subcategory=use_subcategory, notes=notes, raw_source=raw_source
    )

with driver.session() as session:
    for index, row in uses_data.iterrows():
        session.write_transaction(create_uses_nodes,row['id'], row['species'], row['tribe'], row['source'], row['pageno'], row['use_category'], row['use_subcategory'], row['notes'], row['rawsource'])
##############################################################
#nodes for species

def create_species_nodes(tx, id,name, common_names, usda_code, family, family_apg):
    tx.run(
        "CREATE (:Species {id: $id,name: $name, common_names: $common_names, usda_code: $usda_code, family: $family, family_apg: $family_apg})",
        id=id,name=name, common_names=common_names, usda_code=usda_code, family=family, family_apg=family_apg
    )

with driver.session() as session:
    for index, row in species_fam_data.iterrows():
        session.write_transaction(create_species_nodes, row['id'], row['name'], row['common_names'], row['family'], row['usda_code'],row['family_apg'])

##############################################################
#nodes for use_category
def create_categories_nodes(tx, _id, name):
    tx.run(
        "CREATE (:UseCategory {id: $id, name: $name})",
        id=_id, name=name
    )

with driver.session() as session:
    for index, row in use_categories_data.iterrows():
        session.write_transaction(create_categories_nodes, row['id'], row['name'])
######################################################################

#node for use_subcategory
def create_usesubcategory_nodes(tx, _id, parent, name):
    tx.run(
        "CREATE (:UseSubcategory {id: $id, parent: $parent, name: $name})",
        id=_id, parent=parent, name=name
    )

with driver.session() as session:
    for index, row in use_subcategories_data.iterrows():
        session.write_transaction(create_usesubcategory_nodes, row['id'], row['parent'], row['name'])
	
	

```

```python
def create_compound_structure_relationship(tx):
    tx.run(
        """
        MATCH (c:Compound), (s:Structure)
        WHERE c.np_id = s.np_id
        CREATE (c)-[:Has_structure]->(s)
        """
    )
# Create relationships
with driver.session() as session:
    session.write_transaction(create_compound_structure_relationship)
################################################################
def create_usage_species_relationship(tx):
    tx.run(
        """
        MATCH (u:Usage), (sp:Species)
        WHERE u.species = sp.id
        CREATE (u)-[:Has_species_family]->(sp)
        """
    )
# Create relationships
with driver.session() as session:
    session.write_transaction(create_usage_species_relationship)
    
###############################################
def create_usage_category_relationship(tx):
    tx.run(
        """
        MATCH (u:Usage), (uc:UseCategory)
        WHERE u.use_category = uc.id
        CREATE (u)-[:Has_usage_category]->(uc)
        """
    )
# Create relationships
with driver.session() as session:
    session.write_transaction(create_usage_category_relationship)
    
####################################################################################

def create_usage_subcategory_relationship(tx):
    tx.run(
        """
        MATCH (u:Usage), (usc:UseSubcategory)
        WHERE u.use_subcategory = usc.id
        CREATE (u)-[:Has_usage_sub_category]->(usc)
        """
    )
# Create relationships
with driver.session() as session:
    session.write_transaction(create_usage_subcategory_relationship)
    
#########################################################################################

def create_usage_category__subcategory_relationship(tx):
    tx.run(
        """
        MATCH (uc:UseCategory), (usc:UseSubcategory)
        WHERE uc.id = usc.parent
        CREATE (uc)-[:Used_for]->(usc)
        """
    )
# Create relationships
with driver.session() as session:
    session.write_transaction(create_usage_category__subcategory_relationship)
```
