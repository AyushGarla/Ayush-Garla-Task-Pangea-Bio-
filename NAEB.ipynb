species
--

```python
species_fam_data = pd.read_csv("F:\\MS\\SRH study\\IV sem\\INT\\Datasets 2023\\naeb\\data\\naeb_dump\\species.csv")
species_fam_data
```

```python
species_fam_data.isnull().sum()
```

```python
for val in species_fam_data:
    species_fam_data[val].fillna('Missing',inplace=True)
species_fam_data.head()
```

```python
# Checking for duplicate values
duplicates = species_fam_data[species_fam_data.duplicated(subset='id', keep=False)]
duplicate_count = duplicates['id'].nunique()  # Count unique duplicate np_id values

if not duplicates.empty:
    print("Duplicate np_id values found:")
    print(duplicates)
    print(f"Number of duplicate np_id values: {duplicate_count}")
else:
    print("No duplicate np_id values found.")

# Removing duplicate values
species_fam_data = species_fam_data.drop_duplicates(subset='id', keep='first')

species_fam_data
```

```python
species_fam_data.columns = [col.strip().lower() for col in species_fam_data.columns] #keeping column names standard
```

```python
import psycopg2
from sqlalchemy import create_engine
import pandas as pd

connection = psycopg2.connect(
    host="localhost",
    database="compound",
    user="postgres",
    password="123456"
)

engine = create_engine('postgresql://postgres:123456@localhost/compound')

uses_data.to_sql('naeb_uses_table', engine, if_exists='replace', index=False)
species_fam_data.to_sql('naeb_species_fam_table', engine, if_exists='replace', index=False)
tribes_data.to_sql('naeb_tribes_table', engine, if_exists='replace', index=False)
use_categories_data.to_sql('naeb_use_categories_table', engine, if_exists='replace', index=False)
use_subcategories_data.to_sql('naeb_use_subcategories_table', engine, if_exists='replace', index=False)
sources_book_data.to_sql('naeb_sources_book_table', engine, if_exists='replace', index=False)

connection.close()

print("DataFrame successfully written to PostgreSQL table.")
```
